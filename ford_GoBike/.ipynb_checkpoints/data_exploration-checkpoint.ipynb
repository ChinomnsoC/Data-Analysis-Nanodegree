```python
# Remove outliers in birth year column 
elderly_id = ford_df[ford_df['member_birth_year']<1922].index
ford_df = ford_df.drop(index=elderly_id)
# Test
ford_df.member_birth_year.describe()
```

```python
# Create `age` column.
ford_df['age'] = 2022 - ford_df['member_birth_year']
ford_df.age.describe()
```

The age range is now between 21 and 95 years old.

```python
#Create a distance column from the longitude and latitude columns.
def calculate_distance_km(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    longitude_difference = lon2 - lon1
    latitude_difference = lat2 - lat1

    a = np.sin(latitude_difference/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(longitude_difference/2.0)**2

    c = 2 * np.arcsin(np.sqrt(a))
    km = 6367 * c
    return km
```

```python
#Create a distance column from the longitude and latitude columns.
ford_df['distance_km'] = calculate_distance_km(ford_df['start_station_latitude'],ford_df['start_station_longitude'],ford_df['end_station_latitude'],ford_df['end_station_longitude'])
# Test
ford_df.distance_km.describe()
```

It may be best to approximate the values to two decimal places.

```python
ford_df['distance_km'] = ford_df['distance_km'].round(decimals = 2)
ford_df.distance_km.head()
```

```python
# Create new column for day of week, and another for time of the day (morning, afternoon, night, Sunday-Saturday).

# create new column (day of week)
ford_df['start_day_of_week'] = ford_df['start_time'].dt.day_name()
ford_df['end_day_of_week'] = ford_df['end_time'].dt.day_name()

ford_df['start_day_of_week'] = pd.Categorical(ford_df['start_day_of_week'], 
                                       categories=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'], ordered=True)

# create new column (time of the day, hour)
ford_df['start_hour'] = ford_df['start_time'].apply(lambda time: time.hour)
ford_df['start_time_of_day'] = 'morning'
ford_df['start_time_of_day'][(ford_df['start_hour'] >= 12) & (ford_df['start_hour'] <= 17)] = 'afternoon'
ford_df['start_time_of_day'][(ford_df['start_hour'] >= 18) & (ford_df['start_hour'] <= 23)] = 'night'
ford_df.head()
```

```python
# Create a duration minute column
ford_df['duration_minute'] = ford_df['duration_sec']/60
```

```python
# Check for duplicates in the dataset.
def check_and_remove_duplicates_in(ford_df): 
    df_without_duplicates = ford_df.drop_duplicates()
    len_dup = len(ford_df)-len(df_without_duplicates)
    if len_dup > 0 : 
        print (f'This dataset has {len_dup} duplicate values')
        print ('Removing duplicate values......')
        ford_df = ford_df.drop_duplicates()
    else : 
        print ('There are no duplicate rows in this dataset')
check_and_remove_duplicates_in(ford_df)
```

```python
# Drop columns not relevant to the analysis
columns_to_be_deleted = ['start_time', 'end_time', 
       'start_station_id', 'end_station_id',
       'start_station_name', 'start_station_latitude',
       'start_station_longitude', 'end_station_name',
       'end_station_latitude', 'end_station_longitude',
       'member_birth_year', 'bike_share_for_all_trip' ]

ford_df = ford_df.drop(columns= columns_to_be_deleted, axis=1)
ford_df.info()
```

```python
# Delete rows with missing data.
ford_df = ford_df.dropna()
ford_df.info()
```

## Multivariate Exploration


```python
# Number of Hourly Bike Rides in a Day by Subscribers and Customers
plt.figure(figsize=(10,7))
# ford_df['start_day_of_week'] = pd.Categorical(ford_df['start_day_of_week'], 
#                                        categories=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'], ordered=True)

# left plot: subscriber
plt.subplot(1, 2, 1)
num_of_rides_subscribers = ford_df.query('user_type == "Subscriber"').groupby(['start_day_of_week', 
                                                                           'start_hour']).size().astype(int)
num_of_rides_subscribers = num_of_rides_subscribers.reset_index(name = 'count') 
num_of_rides_subscribers = num_of_rides_subscribers.pivot(index = 'start_hour', columns = 'start_day_of_week', 
                                      values = 'count')
ax1 = sb.heatmap(num_of_rides_subscribers, cmap ='crest') 
plt.xlabel('')
plt.title('Subscriber')
ax1.set_ylim(23.5 + 0.5, 0.5 - 0.5)  #ax1.get_ylim() -- (23.5, 0.5)

# right plot: customer
plt.subplot(1, 2, 2)
num_of_rides_customers = ford_df.query('user_type == "Customer"').groupby(['start_day_of_week', 
                                                                         'start_hour']).size().astype(int)
num_of_rides_customers = num_of_rides_customers.reset_index(name = 'count') 
num_of_rides_customers = num_of_rides_customers.pivot(index = 'start_hour', columns = 'start_day_of_week', 
                                      values = 'count')
ax2 = sb.heatmap(num_of_rides_customers, cmap ='crest', cbar_kws = {'label' : 'count'}) 
plt.xlabel('')
plt.ylabel('')
plt.title('Customer')
plt.suptitle('Number of Hourly Bike Rides in a Day by Subscribers and Customers ', fontsize=20)
plt.ylim(ax1.get_ylim());    # set y-axis limits to be same as left plot
```

> - The main interest of this research was to explore the distance covered by each person, and how this may be closely related with other features such as age, gender, bike id, user type, day of the week.

> - To do this, distance column was created from the longitude and latitude columns, and values within the column were rounded up to 2dp. Also, the duration_sec was closely looked at because distance is often associated with time.

> - It was seen that 99% of the cyclers rode for less than an hour. Also, only 1% of the population cycled for more than 5.05km. The people in this latter group were referred to as the ambitious bikers throughout the analysis.

The weekend (Saturday and Sunday) is less likely to be chosen as a day to start a bike trip.

> - **In terms of Gender:** Across all genders, older people tend to ride for shorter distances. Also, more people from the `Other` gender category cycle for shorter distances. It was interesting to see that there are more ambitious female bikers between the age of 60 - 70, than there are for either of the two other genders.

> - **Comparing Customers to Subscribers:** During the weekend, there are more customer riders than there are subscriber riders.
